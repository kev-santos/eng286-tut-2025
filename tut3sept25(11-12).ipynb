# For loops

Note: in a for loop, we use something called a "dummy" or "local" variable. For example, in the question below, ``word`` is a dummy variable. It stands in for whatever element in the list the ``for`` loop is currently on. 

To illustrate this, study the following code blocks. They all do the exact same thing!

```python
list_of_strings = ["red", "orange", "yellow", "green", "blue", "indigo", "violet"]

for word in list_of_strings:
    print(word)
```

```python
list_of_strings = ["red", "orange", "yellow", "green", "blue", "indigo", "violet"]

for colour in list_of_strings:
    print(colour)
```

### Question 10
Write a comment explaining what each line of code is doing in the cell below. (Replace the text after the `#`.) Compare with your neighbour. (This piece of code is an important excerpt from lecture!)

```python
sentence = "I celebrate myself, and sing myself, and what I assume you shall assume, for every atom belonging to me as good belongs to you."
# create a variable called sentence containing a string

sentence_words = sentence.split() # Using the split method to split the string in "sentence" into a list of words
unique_words = [] # Creating an empty list and putting it in the variable unique_words

for word in sentence_words: # iterate through the list sentence_words, labelling each element as "word"
    if word not in unique_words: # comment
        unique_words.append(word) # comment
```

```python
sentence = "I celebrate myself, and sing myself, and what I assume you shall assume, for every atom belonging to me as good belongs to you."
sentence_words = sentence.split()
lowercase_words = []

for word in sentence_words:
    lower_word = word.lower()
    lowercase_words.append(lower_word)

print(lowercase_words)
```
